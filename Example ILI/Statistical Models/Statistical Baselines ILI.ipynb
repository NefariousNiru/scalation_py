```python
import csv
import itertools
import math
import os
import time
from datetime import datetime

if "SARIMAX" not in globals():
    raise RuntimeError("`SARIMAX` is not defined. Import it first.")
if "dotdict" not in globals():
    raise RuntimeError("`dotdict` is not defined. Import it first.")

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
out_dir = f"sarimax_search_{timestamp}"
os.makedirs(out_dir, exist_ok=True)

stage1_csv = os.path.join(out_dir, "stage1_results.csv")
stage2_csv = os.path.join(out_dir, "stage2_results.csv")
best_csv = os.path.join(out_dir, "best_qof.csv")
best_txt = os.path.join(out_dir, "best_result.txt")
log_txt = os.path.join(out_dir, "search_log.txt")


def log(msg):
    print(msg, flush=True)
    with open(log_txt, "a", encoding="utf-8") as f:
        f.write(msg + "\n")


def build_args(
    *,
    p=2,
    d=0,
    q=1,
    P=1,
    D=0,
    Q=0,
    s=52,
    rc=26,
    fourier_order=3,
    bump_mu=6.0,
    bump_s=3.0,
    bump_nu=4.0,
    use_trend_exog=True,
):
    args = dotdict()
    args["data_path"] = "../../datasets/ili/national_illness_24.csv"
    args["dataset"] = "ILI"
    args["date_col"] = "date"
    args["target"] = "ILITOTAL"
    args["training_ratio"] = 80
    args["normalization"] = "log"

    args["p"] = p
    args["d"] = d
    args["q"] = q
    args["P"] = P
    args["D"] = D
    args["Q"] = Q
    args["s"] = s

    args["forecast_type"] = "point"
    args["plot_scope_scale"] = None
    args["plot_eda"] = False
    args["qof_calculation_mode"] = "single_horizon"
    args["trend"] = "n"
    args["horizons"] = list(range(1, 13))

    args["rc"] = rc
    args["fit_method"] = "lbfgs"
    args["optimizer_maxiter"] = 100
    args["same_n_samples"] = True
    args["features"] = "S"
    args["skip_insample"] = None
    args["modeling_approach"] = "joint"
    args["debugging"] = False

    args["clip_dates"] = True
    args["start_date"] = "2020-04-13"
    args["end_date"] = "2022-04-04"

    args["qof_to_display"] = [
        "h",
        "n",
        "MSE Normalized",
        "MAE Normalized",
        "MAE Original",
        "sMAPE Original",
    ]

    args["use_fourier_exog"] = True
    args["fourier_order"] = fourier_order
    args["use_bump_exog"] = True
    args["bump_mu"] = bump_mu
    args["bump_s"] = bump_s
    args["bump_nu"] = bump_nu
    args["use_trend_exog"] = use_trend_exog

    args["include_provider_exog"] = False
    args["include_patient_exog"] = False

    return args


def safe_float(x):
    try:
        return float(x)
    except Exception:
        return math.nan


def extract_avg_metrics(model):
    rows = model.qof.to_dict("records")

    avg_row = None
    horizon_rows = []

    for row in rows:
        if str(row.get("h")) == "Avg.":
            avg_row = row
        else:
            try:
                float(row.get("h"))
                horizon_rows.append(row)
            except Exception:
                pass

    if avg_row is not None:
        avg_mae = safe_float(avg_row.get("MAE Original"))
        avg_smape = safe_float(avg_row.get("sMAPE Original"))
    else:
        maes = [safe_float(r.get("MAE Original")) for r in horizon_rows]
        smapes = [safe_float(r.get("sMAPE Original")) for r in horizon_rows]
        maes = [x for x in maes if not math.isnan(x)]
        smapes = [x for x in smapes if not math.isnan(x)]
        avg_mae = sum(maes) / len(maes) if maes else math.nan
        avg_smape = sum(smapes) / len(smapes) if smapes else math.nan

    return avg_mae, avg_smape, rows


def write_csv(path, rows):
    if not rows:
        return
    keys = set()
    for r in rows:
        keys.update(r.keys())
    keys = list(keys)
    with open(path, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=keys)
        writer.writeheader()
        writer.writerows(rows)


def save_best_qof(path, qof_rows):
    if not qof_rows:
        return
    keys = set()
    for r in qof_rows:
        keys.update(r.keys())
    keys = list(keys)
    with open(path, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=keys)
        writer.writeheader()
        writer.writerows(qof_rows)


def run_one(config, label, idx, total):
    log(f"\n[{label} {idx}/{total}] {config}")
    args = build_args(**config)

    start = time.time()
    model = SARIMAX(args)
    _ = model.trainNtest()
    elapsed = time.time() - start

    avg_mae, avg_smape, qof_rows = extract_avg_metrics(model)

    result = {
        **config,
        "avg_mae_original": avg_mae,
        "avg_smape_original": avg_smape,
        "elapsed_sec": elapsed,
        "status": "ok",
    }

    log(
        f" -> sMAPE={avg_smape:.4f}, "
        f"MAE={avg_mae:.4f}, "
        f"time={elapsed:.1f}s"
    )
    return result, qof_rows


# -------------------------
# safer 66-run overnight search
# Stage 1: 12 runs
# Stage 2: 54 runs
# -------------------------

stage1_space = {
    "p": [1, 2, 3],
    "q": [1, 2],
    "rc": [13, 26],
}

stage2_space = {
    "fourier_order": [2, 3, 4],
    "bump_mu": [5.0, 6.0, 7.0],
    "bump_s": [2.0, 3.0, 4.0],
    "use_trend_exog": [True, False],
}

stage1_keys = list(stage1_space.keys())
stage1_combos = list(itertools.product(*[stage1_space[k] for k in stage1_keys]))

stage2_keys = list(stage2_space.keys())
stage2_combos = list(itertools.product(*[stage2_space[k] for k in stage2_keys]))

log(f"Output dir: {out_dir}")
log(f"Stage 1 runs: {len(stage1_combos)}")
log(f"Stage 2 runs: {len(stage2_combos)}")
log(f"Total runs: {len(stage1_combos) + len(stage2_combos)}")

# -------------------------
# Stage 1
# -------------------------
stage1_results = []
best_stage1 = None
best_stage1_qof = None

for idx, combo in enumerate(stage1_combos, start=1):
    config = {
        "p": combo[0],
        "q": combo[1],
        "rc": combo[2],
    }

    try:
        result, qof_rows = run_one(config, "Stage1", idx, len(stage1_combos))
    except Exception as e:
        result = {
            **config,
            "avg_mae_original": math.nan,
            "avg_smape_original": math.nan,
            "elapsed_sec": math.nan,
            "status": "failed",
            "error": str(e),
        }
        qof_rows = None
        log(f" -> FAILED: {e}")

    stage1_results.append(result)
    write_csv(stage1_csv, stage1_results)

    if result["status"] == "ok":
        if (
            best_stage1 is None
            or result["avg_smape_original"] < best_stage1["avg_smape_original"]
            or (
                result["avg_smape_original"] == best_stage1["avg_smape_original"]
                and result["avg_mae_original"] < best_stage1["avg_mae_original"]
            )
        ):
            best_stage1 = result
            best_stage1_qof = qof_rows
            save_best_qof(best_csv, best_stage1_qof)

if best_stage1 is None:
    raise RuntimeError("No successful Stage 1 runs.")

log(f"\nBest Stage 1: {best_stage1}")

best_backbone = {
    "p": int(best_stage1["p"]),
    "q": int(best_stage1["q"]),
    "rc": int(best_stage1["rc"]),
}

# -------------------------
# Stage 2
# -------------------------
stage2_results = []
best_overall = best_stage1
best_overall_qof = best_stage1_qof

for idx, combo in enumerate(stage2_combos, start=1):
    config = {
        **best_backbone,
        "fourier_order": combo[0],
        "bump_mu": combo[1],
        "bump_s": combo[2],
        "use_trend_exog": combo[3],
    }

    try:
        result, qof_rows = run_one(config, "Stage2", idx, len(stage2_combos))
    except Exception as e:
        result = {
            **config,
            "avg_mae_original": math.nan,
            "avg_smape_original": math.nan,
            "elapsed_sec": math.nan,
            "status": "failed",
            "error": str(e),
        }
        qof_rows = None
        log(f" -> FAILED: {e}")

    stage2_results.append(result)
    write_csv(stage2_csv, stage2_results)

    if result["status"] == "ok":
        if (
            best_overall is None
            or result["avg_smape_original"] < best_overall["avg_smape_original"]
            or (
                result["avg_smape_original"] == best_overall["avg_smape_original"]
                and result["avg_mae_original"] < best_overall["avg_mae_original"]
            )
        ):
            best_overall = result
            best_overall_qof = qof_rows
            save_best_qof(best_csv, best_overall_qof)

with open(best_txt, "w", encoding="utf-8") as f:
    for k, v in best_overall.items():
        f.write(f"{k}: {v}\n")

log("\nSearch complete.")
log(f"Best overall: {best_overall}")
log(f"Stage 1 CSV: {stage1_csv}")
log(f"Stage 2 CSV: {stage2_csv}")
log(f"Best QoF CSV: {best_csv}")
log(f"Best summary TXT: {best_txt}")

print("\nFINAL BEST CONFIG")
print(best_overall)
```
